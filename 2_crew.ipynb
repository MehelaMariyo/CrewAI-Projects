```python
from lib2to3.fixes.fix_input import context

from dotenv import load_dotenv
from scripts.regsetup import description
from sympy.physics.units import temperature

load_dotenv()
```

```python
from crewai import LLM

llm = LLM(
    model="gemini/gemini-3.1-flash-lite-preview",
    temperature=0.1
)
```

```python
from crewai import Agent, Task, Crew

research_agent = Agent(
    role="Research Specialist",
    goal="Research interesting facts about the topic: {topic}",
    backstory="You are an expert at finding relevant and factual data",
    llm=llm,
    verbose=True
)

writer_agent = Agent(
    role="Creative Writer",
    goal="Write a short blog summary using research",
    backstory="You are skilled at writing engaging summaries based on provided content",
    llm=llm,
    verbose=True
)
```

```python
task1 = Task(
    description="Find 3-5 interesting and most recent facts about the topic: {topic}",
    agent=research_agent,
    expected_output="A bullet list of 3-5 facts"
)

task2 = Task(
    description="Write a 100-word blog post summary about the topic: {topic} using the facts form the research",
    agent=writer_agent,
    expected_output="A blog post summary",
    context=[task1]
)
```

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[task1, task2],
    verbose=True
)
```

```python
crew.kickoff(inputs={"topic":"The future of Iran and Israel/USA war"})
```
